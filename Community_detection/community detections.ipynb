```python
import pandas as pd
from Community_detection.community_graph_script import plot_community_graph
```

# Ukraine before

```python
df_ner_ukraine_before = pd.read_csv("../NER and ED/Results/Ukraine_before_with_NER.csv")
df_entities_ukraine_before = pd.read_csv("../NER and ED/Results/Ukraine_before_top_40_entities.csv")
#drop rows wehere word =  U B
df_entities_ukraine_before = df_entities_ukraine_before[~df_entities_ukraine_before.Word.isin(["U", "B", "N", "19", "G", "S"])]
```

```python
suptitle = "Ukraine before conflict\n Co-occurrence in Same Sentence Relationship Graph"
title = "Nodes represent entities. Edges represent co-occurrence within the same article.\nNodes size indicates the node strength.\nEdge width indicates the frequency of co-occurrence. Spring Layout"

plot_community_graph(df_ner_ukraine_before, df_entities_ukraine_before, suptitle=suptitle, title=title, nodes_displayed= 25)
```
